```python
df_region = Demanda[['anio','region','demanda_TWh']][Demanda['anio']<2019].groupby(by=['anio','region']).sum()
df_r = pd.DataFrame()
df_r['region'] = df_region['demanda_TWh'].index.droplevel()
df_r['demanda_TWh'] = df_region.values
df_r.groupby(by=['region']).median()
```

```python
df_r.groupby(by=['region']).median().values
```

```python
fig, ax = plt.subplots(figsize=(9, 6), subplot_kw=dict(aspect="equal"))

data = df_r.groupby(by=['region']).median().values
ingredients = df_r.region.unique()


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} TWh)".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          title="Regiones",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=10, weight="bold")

ax.set_title("Demanda por Región")

plt.savefig('demanda_reg.png')
```

```python
import statsmodels.api as sm
```

```python
def plotProcessX(n_samples, rho=0):
    x = w = n_samples['demanda_TWh']
    for t in range(len(n_samples)):
        x[t] = rho * x[t-1] + w[t]

    with plt.style.context('bmh'):  
        plt.figure(figsize=(10, 3))
        plt.plot(x)
        plt.xticks(rotation=90)
        plt.title("Rho {}\n Dickey-Fuller p-value: {}".format(rho, round(sm.tsa.stattools.adfuller(x)[1], 3)))
        plt.savefig('Dickey-Fuller_Rho'+str(rho)+'.png')
```

```python
linea = Demanda[['indice_tiempo','demanda_TWh']].groupby(by=['indice_tiempo']).sum()
for rho in [0, 0.6, 0.9, 1]:
    plotProcessX(linea, rho=rho)
```

* p-value > 0.05: Fail to reject the null hypothesis (H0), the data has a unit root and is non-stationary.
* p-value <= 0.05: Reject the null hypothesis (H0), the data does not have a unit root and is stationary.
